In this lab we will look at an inclusive search for the standard model Higgs boson in proton-proton colissions at ‚àö s = 13 TeV at the Large Hadron Collider (LHC) using simulated data [A. Schuy]. Proton-proton collisions produce an avalanche of varying particles (i.e. quarks and gluons) which rearrange into hadrons. These hadrons along with their decay products are observed via large particle detectors which in turn are studied by scientists. <br>
<br> 
Higgs boson are produced with large transverse momentum (pT) and decay into a bottom quark-antriquark pair. That said, the Higgs candidates are able to be reconstructed as large-radius jets using Calorimeters. With the training sets provided for anlaysjs, each sample contains 14 features: ‚Äòpt', 'eta', 'phi', 'mass', 'ee2', 'ee3', 'd2', 'angularity', 't1', 't2', 't3', 't21', 't32', 'KtDeltaR' [Eur. Phys. J. C 79 (2019) 836]. We will briefly go over and describe what each feature represents. <br>
<br>

At the LHC there are two main detectors that measure collimated particles, which are also known as jets. To understand how our data was obtain it would beneficial to understand what happens inside these detectors. In principaility, both detector share the same basic design and are essentially considered the same. To be able to measure the tracks of charged particles there is an inner detector that is close to the beam line. In order to measure the energy of electrons and photons an elctromagnetic calorimteter (ecal) follows the inner detector. A hadronic calorimeter (hcal) measures the energies of neutrons and protons. Externally, there is a muon detector. The inner detector contains a tracking system that is composed of different parts. On the outside there is a semiconductor tracker (made of silicon), followed by a trasition radiation tracker. Together these trackers combine together and generate a clear image of what charged partciles were produced by the collision and which direction they went. 
<br>
<br>
The LHC produces about 1 billion events per second. This is an incredible amount of data that needs to be reduced from a billion events to 100 events in order to be able to be recorded and analyzed. This reduction is done via the triggering system. A low-level hardware trigger searches for electrons or photons. Higher-level triggers involve processing with software, like the jet finding algorithm. If an event is not found by a trigger then it is lost forever. 
<br>
<br>
In the inner detector, curvature of tracks is used to distinguish charged particles like electrons, positrons and pions, and to aid in measuring their 3-momentum. 
<br>
<br>
A jet is a collection of particles that travel in the same direction in a detector. A jet allows for a calculation done at the patron level (pions and protons).To reduce systematic uncertrainty of jet measurements, a jet algorithm should be locan in the detector and not pulling particles from far away. The simplest way to define a jet is to draw a cone of size  ùëÖ=(‚àö(ŒîùúÇ)2+Œî(ùúô)2)  around some particle and include everything in that cone as the jet.


### 14 Features Description: 
Jet Kinematics: <br>
'pt': jet transverse momentum of collision in xy plane <br>
'eta': jet pseudo rapidity. Angle of partcile in zy plane <br>
'mass': jet mass of object/system that is independent of overall motion of system <br>
<br>
Subjettiness: Measures N-prone sub-structure <br>
't1': 1-subjettiness<br>
't2': 2-subjettiness<br>
't3': 3-subjettiness<br>
't21': 21-subjettiness<br>
't32': 32-subjettiness<br>
<br>
Cluster Sequence: <br>
'angularity': Orientation of one feature to another at referenced angle <br>
<br>
'KtDeltaR':  delta R of two subjets within large-R jet<br>
<br>
Energy Correlation Functions: <br>
'ee2': 2 point <br>
'ee3':  3 point <br>
'd2': ee3/ee2 ratio <br>
